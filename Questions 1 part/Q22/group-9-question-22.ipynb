```python
import pandas as pd

swift_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/swift_currency_tracker_all_reports.csv')

swift_df = swift_df.dropna(
    subset=['currency_or_economy', 'rmb_global_rank']
)

swift_df['rmb_global_rank'] = pd.to_numeric(
    swift_df['rmb_global_rank'],
    errors='coerce'
)

top5 = swift_df[
    swift_df['rmb_global_rank'] <= 5
]

result = (
    top5.groupby('currency_or_economy')
    .size()
    .reset_index(name='Times_in_Top_5')
    .sort_values(
        by='Times_in_Top_5',
        ascending=False
    )
)

result.to_csv(
    'currencies_most_often_in_top5.csv',
    index=False
)

print(result)
```
