```python
import pandas as pd

oil_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/Brent Oil.csv')
gold_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/Gold 100years.csv')
silver_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/silver 100 years.csv')

oil_df['Date'] = pd.to_datetime(oil_df['Date'], dayfirst=True)
gold_df['Date'] = pd.to_datetime(gold_df['Date'], dayfirst=True)
silver_df['Date'] = pd.to_datetime(silver_df['Date'], dayfirst=True)

oil_df['Year'] = oil_df['Date'].dt.year
gold_df['Year'] = gold_df['Date'].dt.year
silver_df['Year'] = silver_df['Date'].dt.year

oil_yearly = oil_df.groupby('Year')['Value'].mean().reset_index()
gold_yearly = gold_df.groupby('Year')['Value'].mean().reset_index()
silver_yearly = silver_df.groupby('Year')['Value'].mean().reset_index()

oil_yearly['Oil_Change_%'] = oil_yearly['Value'].pct_change() * 100
gold_yearly['Gold_Change_%'] = gold_yearly['Value'].pct_change() * 100
silver_yearly['Silver_Change_%'] = silver_yearly['Value'].pct_change() * 100

strong_year = oil_yearly.loc[
    oil_yearly['Oil_Change_%'].abs().idxmax(),
    'Year'
]

gold_change = gold_yearly.loc[
    gold_yearly['Year'] == strong_year,
    'Gold_Change_%'
].values[0]

silver_change = silver_yearly.loc[
    silver_yearly['Year'] == strong_year,
    'Silver_Change_%'
].values[0]

oil_change = oil_yearly.loc[
    oil_yearly['Year'] == strong_year,
    'Oil_Change_%'
].values[0]

result = pd.DataFrame({
    'Year': [strong_year],
    'Oil_Change_%': [round(oil_change, 2)],
    'Gold_Change_%': [round(gold_change, 2)],
    'Silver_Change_%': [round(silver_change, 2)]
})

result.to_csv(
    'oil_gold_silver_movement_comparison.csv',
    index=False
)

print(result)
```
