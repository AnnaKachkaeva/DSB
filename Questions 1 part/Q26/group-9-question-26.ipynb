```python
import pandas as pd

oil_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/Brent Oil.csv')

oil_df['Date'] = pd.to_datetime(
    oil_df['Date'],
    dayfirst=True
)

oil_df['Value'] = pd.to_numeric(
    oil_df['Value'],
    errors='coerce'
)

latest_row = oil_df.sort_values('Date').iloc[-1]

latest_value = latest_row['Value']
long_term_average = oil_df['Value'].mean()

status = (
    'Above Long-Term Average'
    if latest_value > long_term_average
    else 'Below Long-Term Average'
)

result = pd.DataFrame({
    'Latest_Brent_Oil_Value': [round(latest_value, 2)],
    'Long_Term_Average': [round(long_term_average, 2)],
    'Comparison': [status]
})

result.to_csv(
    'brent_oil_latest_vs_average.csv',
    index=False
)

print(result)
```
