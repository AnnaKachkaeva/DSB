```python
import pandas as pd
import matplotlib.pyplot as plt

gold_df = pd.read_csv("/kaggle/input/datasets/ladkevinkumar/group9-dsb/Gold 100years.csv")

gold_df["Date"] = pd.to_datetime(gold_df["Date"])

gold_df["Year"] = gold_df["Date"].dt.year

gold_yearly = gold_df.groupby("Year")["Value"].mean().reset_index()

gold_yearly.columns = ["Year", "Average_Gold_Price"]

gold_yearly.to_csv("gold_yearly_average_price.csv", index=False)

plt.figure(figsize=(14,6))

plt.plot(
    gold_yearly["Year"],
    gold_yearly["Average_Gold_Price"],
    linewidth=2
)

plt.title("Gold Prices Over Time")
plt.xlabel("Year")
plt.ylabel("Average Gold Price")
plt.grid(True)

plt.savefig("gold_price_trend.pdf")

plt.show()

print(gold_yearly.head())
```
