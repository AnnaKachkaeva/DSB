```python
import pandas as pd

gold_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/Gold 100years.csv')

gold_df['Date'] = pd.to_datetime(
    gold_df['Date'],
    dayfirst=True
)

gold_df['Value'] = pd.to_numeric(
    gold_df['Value'],
    errors='coerce'
)

latest_row = gold_df.sort_values('Date').iloc[-1]

latest_value = latest_row['Value']
long_term_average = gold_df['Value'].mean()

status = (
    'Above Long-Term Average'
    if latest_value > long_term_average
    else 'Below Long-Term Average'
)

result = pd.DataFrame({
    'Latest_Gold_Value': [round(latest_value, 2)],
    'Long_Term_Average': [round(long_term_average, 2)],
    'Comparison': [status]
})

result.to_csv(
    'gold_latest_vs_average.csv',
    index=False
)

print(result)
```
