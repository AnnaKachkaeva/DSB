```python
# Create a line chart for Silver prices over time and describe the main peaks. 

import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("Preparing data/cleaned data/silver 100 years.csv")

df['Value'] = df['Value'].astype(float)
df["date_date"] = pd.to_datetime(df["date_date"])
df = df.drop(columns=['date_str'])
```

```python
df = df[df['date_date'] > '1968-01-01']
df.plot(figsize=(9, 4),x='date_date', y='Value')
plt.show()
```
