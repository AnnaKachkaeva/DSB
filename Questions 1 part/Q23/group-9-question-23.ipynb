```python
import pandas as pd

swift_df = pd.read_csv('/kaggle/input/datasets/ladkevinkumar/group9-dsb/swift_currency_tracker_all_reports.csv')

offshore_rmb = swift_df[
    swift_df['category'].astype(str).str.contains(
        'offshore',
        case=False,
        na=False
    )
].copy()

offshore_rmb['value'] = pd.to_numeric(
    offshore_rmb['value'],
    errors='coerce'
)

result = (
    offshore_rmb.groupby('currency_or_economy')['value']
    .max()
    .reset_index()
    .sort_values(
        by='value',
        ascending=False
    )
    .head(1)
)

result.columns = [
    'Offshore_RMB_Economy',
    'Highest_Share'
]

result.to_csv(
    'highest_offshore_rmb_economy.csv',
    index=False
)

print(result)
```
