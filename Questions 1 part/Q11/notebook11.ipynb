```python
import pandas as pd
from pathlib import Path

df = pd.read_csv("Preparing data/cleaned data/Gold 100years.csv")

df.info()

df['Value'] = df['Value'].astype(float)

df["date_date"] = pd.to_datetime(df["date_date"])

yearly_avg = (
    df.groupby(df["date_date"].dt.year)["Value"]
      .mean()
      .reset_index(name="avg_value")
)

yearly_avg.sort_values(by=['avg_value'], ascending=False).head(5)
```
