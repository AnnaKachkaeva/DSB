```python
import pandas as pd

silver_df = pd.read_csv("/kaggle/input/datasets/ladkevinkumar/group9-dsb/silver 100 years.csv")
gold_df = pd.read_csv("/kaggle/input/datasets/ladkevinkumar/group9-dsb/Gold 100years.csv")
oil_df = pd.read_csv("/kaggle/input/datasets/ladkevinkumar/group9-dsb/Brent Oil.csv")
```

```python
def convert_date(df):
    df["Date"] = pd.to_datetime(df["Date"])
    df["Month"] = df["Date"].dt.month
    df["Year"] = df["Date"].dt.year
    return df

silver_df = convert_date(silver_df)
gold_df = convert_date(gold_df)
oil_df = convert_date(oil_df)

print("Silver Dataset")
display(silver_df.head())

print("Gold Dataset")
display(gold_df.head())

print("Brent Oil Dataset")
display(oil_df.head())
```

```python
silver_df.to_csv("silver_month_year.csv", index=False)
gold_df.to_csv("gold_month_year.csv", index=False)
oil_df.to_csv("brent_oil_month_year.csv", index=False)
```
